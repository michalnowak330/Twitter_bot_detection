```python
def get_replies(screen_name, no_of_tweets, no_of_replies):

    replies=[]
    reply_to = []
    non_bmp_map = dict.fromkeys(range(0x10000, sys.maxunicode + 1), 0xfffd)

    for full_tweets in tw.Cursor(api.user_timeline, 
                                 screen_name = screen_name, 
                                 include_rts = 'False', 
                                 exclude_replies=True, 
                                 timeout=999999).items(no_of_tweets):

        for tweet in tw.Cursor(api.search, 
                             q='to:'+ screen_name, 
                             result_type='recent', 
                             timeout=999999,
                            tweet_mode='extended').items(no_of_replies):

            if hasattr(tweet, 'in_reply_to_status_id_str'):

                if (tweet.in_reply_to_status_id_str == full_tweets.id_str):
                    replies.append((tweet.author.id, tweet.full_text, tweet.author.name, tweet.in_reply_to_status_id_str))

        reply_to.append((full_tweets.id_str, full_tweets.text.translate(non_bmp_map)))
        
    originals = pd.DataFrame(reply_to, columns = ('id','text'))    
    data = pd.DataFrame(replies, columns=('id', 'text', 'author', 'in_reply_to'))
        
    return(data, originals)

def frame_accounts(ids):
    download = []

    for un_id in ids:
        
        user = api.get_user(un_id)
        download.append((user.id_str,
                            #user.screen_name,
                            user.location,
                            #user.description,
                            user.url,
                            user.followers_count,
                            user.friends_count,
                            user.listed_count,
                            #user.created_at,
                            user.favourites_count,
                            user.verified,
                            #user.statuses_count,
                            #user.lang,
                            user.default_profile,
                            user.default_profile_image,
                            user.has_extended_profile))
                            #user.name))
        
        data = pd.DataFrame(download, columns=('id_str',
                                               'location',
                                               'url',
                                               'followers_count',
                                               'friends_count',
                                               'listed_count',
                                               'favourites_count',
                                               'verified',
                                               'default_profile',
                                               'default_profile_image',
                                               'has_extended_profile'
                                              ))
    return(data)
```
